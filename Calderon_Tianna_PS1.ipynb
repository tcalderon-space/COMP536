```python
#blackbody
#Example
Teff_arr = [5500, 5000, 4500, 4000, 3500]

l_teff = len(Teff_arr)
n_pts = 1000
wl_arr = np.zeros((l_teff, n_pts))
bb_arr = np.zeros((l_teff,n_pts))


for i in range(l_teff):
    wl, bb = sf.blackbody_function(Teff_arr[i])
    wl_arr[i] = wl
    bb_arr[i] = bb
    sf.plot_func(wl_arr[i], bb_arr[i], xmin = 1E-6, xmax = 5E-4, xlabel = "Wavelength - cm", ymin = 0, ymax = 3E14, ylabel = r'Spectral Radiance $kJ/m^3$', flag_y = 1, flag_x = 0, 
       label = Teff_arr[i], Title = "Blackbody Spectra for Values of T_eff", save_fig = False, figname = "Blackbody Spectra for Values of T_eff")
    
```
